```python
import re
def remove_html_tags(text):
  pattern = re.compile("<.*?>")
  return pattern.sub(r"", text)
```

```python
import string
string.punctuation
```

```python
exclude = string.punctuation
```

```python
def remove_punc(text):
  for char in exclude:
    text = text.replace(char , "")
  return text
```

```python
text1 = 'string. with. punctuation?'
```

```python
chat_words = {"u2" : "you too" , "BBL" : "Be Back Later " , "GAL": 'Get A Life'}
```

```python
chat_words
```

```python
def chat_conversation(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
chat_conversation("BBL")
```

```python
# STOPWORDS
```

```python
def remove_stopwords(text):
  new_text = []
  for word in text.split():
    if word in set(stopwords.words("english")):
      new_text.append("")
    else:
      new_text.append(word)
  x = new_text[:]
  new_text.clear()
  return "".join(x)
```

```python
remove_stopwords("probably my all-time favourite movie , a story of selflessness , sacrifice and dedicate ")
```

Stemming


```python
# 1. inflection ===> " in grammar , inflection is the modification of a word
# to express different grametical categories such as tense, case , voicce,aspect , person ,
# number, gender , and mood "

# 2. Stemming ==> Stemming is the process of reducing inflection in words to
# their root forms such as mapping a group of words to the same stem even if the
# stem itself is not valid word in language .
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(text):
  return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample = "walk walks walking walked"
stem_words(sample)
```

```python
# stemming mapping a group of words to their same stem even if the stem itself
# is not valid word in language . So , this is the big issue of stemming

text = "probably my alltime favourite movie a story of selflessness sacrifice and dedication "
print(text)
```

```python
stem_words(text)
```
