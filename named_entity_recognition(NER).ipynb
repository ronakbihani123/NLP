```python
# Named entity recognition (NER) is an NLP technique used to find important names and information in text such
# as people , places , organizations , dates, and amounts and label them with their
# correct type.
```

```python
# # in short:
# NER helps a machine understand "who" , "what" , and "where" from text.
```

```python
# # example (very simple)
# # text:
# # "Rohit sharma lives in mumbai"

# # NER Output:
# # rohit sharma > Person
# # mumbai > Location

# # why do we use NER?
# # we use NER to convert unstructured text into meaningful,  structured data.
# # Main Reason:
# # Automatically extract important information
# reduce manual work in reading documents
# help machines understand context , not just keywords
# make text usable for analytics , ML models and databases
```

```python
# Load NER model
```

```python
import spacy
# load english NLP model
nlp = spacy.load("en_core_web_sm")
```

```python
# ex-1
text = "Barack Obama was borm in Hawaii and became President of the United States."

doc = nlp(text)
for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```

```python
# ex-2
text = """
Google CEO Sundar Pichai announced a $10 billion investment
in India on 15th july 2020."""

doc = nlp(text)
for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```

```python
# exx-3 person & location
text = "Sachin Tendulkar was born in Mumbai"
doc = nlp(text)

for ent in doc.ents:
  print(ent.text, "->", ent.label_)
```

```python
# ex-4 Organization & date
text = "Microsoft was founded in 1975 by Bill Gates"

doc = nlp(text)

for ent in doc.ents:
  print(ent.text, "->", ent.label_)
```

```python
# Visaulizing NER
```

```python
from spacy import displacy
displacy.render(doc , style ="ent", jupyter=True)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download required resources (run once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # Added to resolve the LookupError

review = """Amazing experience, the trainer kuldeep sir and prajjal sir are great.
Environment was good and learning , basic were cleared from start.
Thank to regex I got placed as a web developer and got the opportunity after 12th class."""

# 1. Convert to lowercase
review = review.lower()

# 2. Remove numbers
review = re.sub(r'\d+', '', review)

# 3. Remove punctuation
review = re.sub(r'[^\w\s]', '', review)

# 4. Tokenize
tokens = word_tokenize(review)

# 5. Remove stopwords
stop_words = set(stopwords.words('english'))
clean_tokens = [word for word in tokens if word not in stop_words]

# 6. Join back to sentence
clean_review = " ".join(clean_tokens)

print(clean_review)
```
